```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Lets take four different arrays holding co-ordinates of two points.
x1,y1 and x2,y2.

```python
x1= [1,12,34,55,45,78,5,9,88]
y1 = [23,56,77,75,59,3,66,0,10]
x2 = [3,56,17,5,59,3,56,60,1]
y2 = [3,56,27,95,9,34,6,40,80]
```

we imported matplotlib as plt right?, so plt has some methods which we can use to visualize data.
Here title means title of the plot and we are using scatter plot so plt.scatter() is method to do that. We pass our x1,y1 to thid method.final we add labels. and call show() to print the graph.

```python
plt.title("Scatter plot")
plt.scatter(x1,y1, c = 'k')
plt.scatter(x2,y2, c = 'b')
plt.xlabel("X lable")
plt.ylabel("Y Lable")
plt.show()
```

Here we just used bar chart instead of scatter, so only change you need to do is use plt.bar().
the other arguments are self-descriptive.

```python
plt.bar(x1,y1,color='black',width=10)
plt.bar(x2,y2,color='blue',width=10)
plt.show()
```

Now lets try this on actual data set.
